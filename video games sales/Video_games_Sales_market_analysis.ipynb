##  Market Trend Analysis
This section analyzes how the video game industry has evolved over time, focusing on sales trends, genre popularity, and platform dominance.

```python
import pandas as pd

import matplotlib.pyplot as plt
```

###  Data Loading
The dataset is loaded using pandas and contains information about video games including genre, platform, sales, and critic scores.

```python
df = pd.read_csv("Video_Games_Sales_Cleaned.csv")
```

###  Data Cleaning
The dataset is already cleaned:
- No missing values in important columns
- Release year is already extracted
- Data is ready for analysis

### Yearly Sales Trend


```python
year_sales = df.groupby("release_year")["total_sales"].sum()
plt.plot(year_sales)
plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid(alpha=0.3)  
plt.show()
```

The graph shows total game sales over the years.

- Rapid growth observed from 2000 to 2010  
- Peak sales around mid-2000s  
- Decline after 2015 (possible shift to digital/mobile gaming)

### Cumulative Sales Growth


```python
year_sales.cumsum().plot()
plt.title("Cumulative Sales Growth")
plt.xlabel("Year")
plt.grid(alpha=0.3)  
plt.show()
```

Cumulative sales show continuous growth of the gaming industry over time, indicating increasing market size.

### Genre Trends Over Time


```python
pivot =df.groupby(["release_year","genre"])["total_sales"].sum().unstack()

plt.figure(figsize=(14 ,7))
for col in pivot.columns:
    plt.plot(pivot.index, pivot[col], alpha=0.5,label=col)

plt.title("All Genres Trend")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)  
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.show()
```

- Action and Sports genres dominate most years  
- Shooter genre shows strong growth in later years  
- Some genres rise and fall based on trends

### Platform Trends


```python
console_trend = df.groupby("console")["total_sales"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Console Trend")
plt.xlabel("Console")
plt.ylabel("Total Sales")
plt.show()
```

- Certain consoles dominate specific time periods  
- New console releases lead to spikes in game releases  
- Older platforms decline over time

### Publisher Trends


```python
df.groupby("publisher")["total_sales"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Publisher Trend")
plt.xlabel("Publisher")
plt.ylabel("Total Sales")
plt.show()
```

- Few major publishers dominate the market  
- Companies like EA, Activision, and Ubisoft consistently release popular games  
- Market shows partial monopolistic behavior

### Impact of Critic Scores


```python
plt.scatter(df["critic_score"], df["total_sales"])
plt.title('Critic Score Impact')
plt.xlabel("Critic score")
plt.ylabel("Total Sales")
plt.show()
```

- Weak positive correlation between critic scores and sales  
- High-rated games tend to sell more, but not always  
- Other factors like marketing and brand also influence sales

### Key Market Insights

- Gaming industry experienced rapid expansion in early 2000s  
- Action and Shooter genres drive most revenue  
- Console lifecycle strongly affects game releases  
- Market is dominated by a few major publishers  
- Sales distribution is highly uneven (few hits dominate)

### Market Conclusion

The video game industry has shown strong growth over time with clear dominance of certain genres and platforms. While innovation and new consoles drive growth, the market is highly competitive and concentrated among major publishers. Future trends may shift towards digital and online gaming.
